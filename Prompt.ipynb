## 使用ChatGPT4撰寫，結合先前的程式碼去作出結構完整的新Code

```python
import requests
import json
from pathlib import Path

# 1. 變數、字串定義
API_KEY = "YOUR_API_KEY"  # 請將YOUR_API_KEY替換為您的API密鑰
Role_Set = "engineer"  # 預設的角色設定

# 2. 函式定義
# 生成程式碼, 第一次呼叫
def Post_Prompt(Message):
    response = requests.post(
        'https://api.openai.com/v1/chat/completions',
        headers={
            'Content-Type': 'application/json',
            'Authorization': f'Bearer {API_KEY}'
        },
        json={
            'model': 'gpt-3.5-turbo-0613',
            'messages': [
                {"role": "system", "content": Role_Set},
                {"role": "user", "content": Message}],
        }
    )
    print(response.json())
    return response

# 自訂腳色和訊息
def Post_Prompt_Custom(Message, Role):
    response = requests.post(
        'https://api.openai.com/v1/chat/completions',
        headers={
            'Content-Type': 'application/json',
            'Authorization': f'Bearer {API_KEY}'
        },
        json={
            'model': 'gpt-3.5-turbo-0613',
            'messages': [
                {"role": "system", "content": Role},
                {"role": "user", "content": Message}],
        }
    )
    print(response.json())
    return response

# 自訂腳色、訊息和溫度
def Post_Prompt_Custom_Temp(Message, Role, Temperature):
    response = requests.post(
        'https://api.openai.com/v1/chat/completions',
        headers={
            'Content-Type': 'application/json',
            'Authorization': f'Bearer {API_KEY}'
        },
        json={
            'model': 'gpt-3.5-turbo-0613',
            'messages': [
                {"role": "system", "content": Role},
                {"role": "user", "content": Message}],
            'temperature': Temperature,
        }
    )
    print(response.json())
    return response

# 使用記憶功能的API呼叫
def Post_Prompt_Memory(Message, Assistant, Role):
    response = requests.post(
        'https://api.openai.com/v1/chat/completions',
        headers={
            'Content-Type': 'application/json',
            'Authorization': f'Bearer {API_KEY}'
        },
        json={
            'model': 'gpt-3.5-turbo-0613',
            'messages': [
                {"role": "system", "content": Role},
                {"role": "user", "content": Message},
                {"role": "assistant", "content": Assistant}
            ],
        }
    )
    print(response.json())
    return response
#檢查是否照著格式生成
def check_responses(directory_name):
    # 定義儲存目錄路徑
    storage_dir = Path(f"./Response_from_ChatGPT/{directory_name}")

    # 初始化計數器和列表
    COUNT = 0
    Response_1st_success = []
    Response_1st_failed = []

    # 讀取儲存目錄中的每個檔案
    for filePath in storage_dir.glob("*.json"):
        with open(filePath, "r", encoding="utf8") as file:
            dic_Tactics = json.load(file)

            # 檢查是否可實現
            if "choices" in dic_Tactics.keys():
                if dic_Tactics["choices"][0]["message"]["content"].splitlines()[0] == "[可實現]":
                    COUNT += 1
                    Response_1st_success.append(filePath.stem)  # 使用.stem來獲取不含副檔名的檔案名
                elif dic_Tactics["choices"][0]["message"]["content"].splitlines()[0] == "[Feasible]":
                    COUNT += 1
                    Response_1st_success.append(filePath.stem)
            else:
                Response_1st_failed.append(filePath.stem)

    # 返回結果
    return COUNT, Response_1st_success, Response_1st_failed
# 3. 函式結構（迴圈呼叫API）# 假設的list_a和Response_Success，實際使用時需要根據實際數據結構進行調整
#此處是進行地一次TTPs生成所用的迴圈
def call_api_and_save_responses(dic_:dict,Prompt:str,Role_:str,save_:str):
    Response_Success = []
    #先更新一次List，確認生成過哪些避免重複生成
    count,Response_Success,List_2=check_responses("1st_en")
    for i in dic_:
        # 檢查是否已經處理過這個項目
        TTPs_ID=dic_[i]["ID"]
        if TTPs_ID not in Response_Success:
            try:
                # 假設進行API呼叫
                response = Post_Prompt_Custom(dic_[i][Prompt],Role_)  # 或其他相關函式，依據需求調整
                # 儲存回應
                File_Name=dic_[i]["ID"]
                with open(f"./Response_from_ChatGPT/{save_}/{File_Name}.json", "w+", encoding="utf8") as file_json:
                    json.dump(response.json(), file_json, indent=2, separators=(',', ':'), ensure_ascii=False)
                # 更新Response_Success列表
                count,Response_Success,List_2=check_responses("1st_en")
            except Exception as e:
                print(f"處理項目 {File_Name} 時發生錯誤: {e}")
    return Response_Success
```

```python
import pandas as pd
from pathlib import Path

def create_excel_from_json(directory_name):
    # 定義存儲目錄路徑
    storage_dir = Path(f"./Response_from_ChatGPT/{directory_name}")

    # 初始化存儲數據的列表
    data = []

    # 遍歷目錄中的JSON文件
    for json_path in storage_dir.glob("*.json"):
        with open(json_path, "r", encoding="utf8") as jsonFile:
            dic_ = json.load(jsonFile)

            # 從字典中提取所需的內容
            if "choices" in dic_ and dic_["choices"]:
                content = dic_["choices"][0]["message"]["content"]
                data.append({"File Name": json_path.stem, "Content": content})

    # 創建DataFrame並保存為Excel文件
    df = pd.DataFrame(data)
    excel_path = storage_dir / "output.xlsx"
    df.to_excel(excel_path, index=False)

    return f"Excel file created at {excel_path}"

# 使用函數並打印結果
directory_name_example = "2nd"  # 替換為實際的目錄名稱
result = create_excel_from_json(directory_name_example)
print(result)
```
