```python
import os
from pathlib import Path
import json
COUNT=0
COUNT_FAIL=0
def scanRecurse(baseDir):
    for entry in os.scandir(baseDir):
        if entry.is_file():
            yield entry
        else:
            yield from scanRecurse(entry.path)
            



directory = './Response_from_ChatGPT/1st' # CHANGE THIS
try:
    for item in scanRecurse(directory): 
        filePath = Path(item)
        fileType = filePath.suffix.lower()
        print(filePath)
        print(fileType)
        with open(filePath,'r',encoding="utf8")as fr:
            #print(fr.read())
            dic_Tactics=json.load(fr)
        print(type(dic_Tactics))
        print("choices" in dic_Tactics.keys())
        if "choices" in dic_Tactics.keys():
            COUNT+=1
            Response_1st_sucess.append(filePath.name.split(".",)[0])
        else:
            Response_1st_failed.append(filePath.name.split(".",)[0])
except:
    print("Done")
    COUNT_FAIL+=1
    
print(COUNT)
print(COUNT_FAIL)
```
